# 1

```python
inputs = open("data/1.txt").read().splitlines()

s1 = s2 = 0

position = 50

for inp in inputs:
    if inp[0] == "L":
        for i in range(int(inp[1:])):
            position = position - 1
            position = position % 100
            if position == 0:
                s2 += 1
    else:
        for i in range(int(inp[1:])):
            position = position + 1
            position = position % 100
            if position == 0:
                s2 += 1

    if position == 0:
        s1 += 1

print(s1, s2)
```

# 2

```python
inputs = open("data/2.txt").read()
inputs = [i.split("-") for i in inputs.split(",")]

final1 = set()
final2 = set()

for a, b in inputs:
    c = 2
    while True:
        d = 1
        while True:
            comb = str(d) * c
            if (int(a) <= int(comb) <= int(b)) and c == 2:
                final1.add(int(comb))
            if int(a) <= int(comb) <= int(b):
                final2.add(int(comb))
            if int(comb) > int(b):
                break
            d += 1
        if c > len(b):
            break
        else:
            c += 1

print(sum(final1), sum(final2))
```

# 3

```python
inputs = open("data/3.txt").read().splitlines()

s1 = s2 = 0

for inp in inputs:
    a = max(inp[:-1])
    b = max(inp[inp.index(a) + 1 :])
    s1 += int(a + b)

for inp in inputs:
    num = ""
    for i in range(11, 0, -1):
        num += str(max(inp[:-i]))
        inp = inp[inp.index(max(inp[:-i])) + 1 :]
    num += max(inp)
    s2 += int(num)

print(s1, s2)
```

# 4

```python
inp = read_input("data/4.txt", matrixify=True)

s1 = s2 = 0

all_locs = inp.get_locs("@")
for loc in all_locs:
    if len(set(all_locs).intersection(inp.get_neighbors(loc, diag=True))) < 4:
        s1 += 1

while True:
    to_clear = []
    all_locs = inp.get_locs("@")
    for loc in all_locs:
        if len(set(all_locs).intersection(inp.get_neighbors(loc, diag=True))) < 4:
            s2 += 1
            to_clear.append(loc)
    if len(to_clear) > 0:
        for x in to_clear:
            inp[x] = "."
    else:
        break

print(s1, s2)
```

# 5

```python
inputs = open("data/5.txt").read().splitlines()

s1 = s2 = 0

ranges = inputs[: inputs.index("")]
ranges = [(int(i.split("-")[0]), int(i.split("-")[1])) for i in ranges]
ingredients = inputs[inputs.index("") + 1 :]

for ing in ingredients:
    for a, b in ranges:
        if int(ing) in range(a, b + 1):
            s1 += 1
            break

union_ranges = []
for begin, end in sorted(ranges):
    if union_ranges and union_ranges[-1][1] >= begin - 1:
        union_ranges[-1][1] = max(union_ranges[-1][1], end)
    else:
        union_ranges.append([begin, end])

for a, b in union_ranges:
    s2 += b - a + 1

print(s1, s2)
```

```python
inputs = open("data/6.txt").read().splitlines()

s1 = s2 = 0

numbers = [i.split() for i in inputs[:-1]]
numbers = list(map(list, zip(*numbers)))
operations = inputs[-1].split()

for op, nums in zip(operations, numbers):
    num = int(nums[0])
    for n in nums[1:]:
        if op == "+":
            num += int(n)
        else:
            num *= int(n)
    s1 += num

numbers = []
nums = []
for x in zip(*inputs[:-1]):
    if ("".join(x)).isspace():
        numbers.append(nums)
        nums = []
    else:
        nums.append(int("".join(x)))
numbers.append(nums)

for op, nums in zip(operations, numbers):
    s = nums[0]
    for num in nums[1:]:
        if op == "+":
            s += num
        else:
            s *= num
    s2 += s

print(s1, s2)
```
