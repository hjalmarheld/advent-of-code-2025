```python
inputs = open("data/1.txt").read().splitlines()

s1 = s2 = 0

position = 50

for inp in inputs:
    if inp[0] == "L":
        for i in range(int(inp[1:])):
            position = position - 1
            position = position % 100
            if position == 0:
                s2 += 1
    else:
        for i in range(int(inp[1:])):
            position = position + 1
            position = position % 100
            if position == 0:
                s2 += 1

    if position == 0:
        s1 += 1

print(s1, s2)
```

```python
inputs = open("data/2.txt").read()
inputs = [i.split("-") for i in inputs.split(",")]

final1 = set()
final2 = set()

for a, b in inputs:
    c = 2
    while True:
        d = 1
        while True:
            comb = str(d) * c
            if (int(a) <= int(comb) <= int(b)) and c == 2:
                final1.add(int(comb))
            if int(a) <= int(comb) <= int(b):
                final2.add(int(comb))
            if int(comb) > int(b):
                break
            d += 1
        if c > len(b):
            break
        else:
            c += 1

print(sum(final1), sum(final2))
```

# 4

```python
inp = read_input("data/4.txt", matrixify=True)

s1 = s2 = 0

all_locs = inp.get_locs("@")
for loc in all_locs:
    if len(set(all_locs).intersection(inp.get_neighbors(loc, diag=True))) < 4:
        s1 += 1

while True:
    to_clear = []
    all_locs = inp.get_locs("@")
    for loc in all_locs:
        if len(set(all_locs).intersection(inp.get_neighbors(loc, diag=True))) < 4:
            s2 += 1
            to_clear.append(loc)
    if len(to_clear) > 0:
        for x in to_clear:
            inp[x] = "."
    else:
        break

print(s1, s2)
```

# 5

```python
inputs = open("data/5.txt").read().splitlines()

s1 = s2 = 0

ranges = inputs[: inputs.index("")]
ranges = [(int(i.split("-")[0]), int(i.split("-")[1])) for i in ranges]
ingredients = inputs[inputs.index("") + 1 :]

for ing in ingredients:
    for a, b in ranges:
        if int(ing) in range(a, b + 1):
            s1 += 1
            break

union_ranges = []
for begin, end in sorted(ranges):
    if union_ranges and union_ranges[-1][1] >= begin - 1:
        union_ranges[-1][1] = max(union_ranges[-1][1], end)
    else:
        union_ranges.append([begin, end])

for a, b in union_ranges:
    s2 += b - a + 1

print(s1, s2)
```

```python
inputs = open("data/7.txt").read().splitlines()


def find_locs(string, substring="^"):
    return [i for i in range(len(string)) if string.startswith(substring, i)]


beams = {inputs[0].index("S"): 1}
rows = [row for row in inputs if "^" in row]

s1 = s2 = 0

for row in rows:
    locs = find_locs(row)
    new_beams = defaultdict(int)
    for beam, count in beams.items():
        if beam in locs:
            s1 += 1
            new_beams[beam - 1] += count
            new_beams[beam + 1] += count
        else:
            new_beams[beam] += count
    beams = new_beams

s2 = sum(beams.values())
```

# 8

```python
inputs = open("data/8.txt").read().splitlines()
inputs = [tuple(map(int, i.split(","))) for i in inputs]

dists = {
    (a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2 + (a[2] - b[2]) ** 2: (a, b)
    for a, b in combinations(inputs, 2)
}

connected = 0
groups = [set([inp]) for inp in inputs]

for _, (a, b) in sorted(dists.items()):
    for group in groups:
        if a in group:
            a_group = group
        if b in group:
            b_group = group

    if a_group != b_group:
        a_group |= b_group
        groups.remove(b_group)

    connected += 1

    if connected == 1000:
        group_sizes = sorted([len(g) for g in groups], reverse=True)
        s1 = group_sizes[0] * group_sizes[1] * group_sizes[2]
    if len(groups) == 1:
        s2 = a[0] * b[0]
        break

print(s1, s2)
```

```python
points = [tuple(map(int, i.split(","))) for i in open("data/9.txt").read().splitlines()]
line_centres = [
    ((x1 + x2) / 2, (y1 + y2) / 2) for (x1, y1), (x2, y2) in zip(points, points[1:])
]

s1 = s2 = 0

for a, b in combinations(points, 2):
    min_x, max_x = sorted((a[0], b[0]))
    min_y, max_y = sorted((a[1], b[1]))

    area = (max_x - min_x + 1) * (max_y - min_y + 1)
    s1 = max((s1, area))

    if area <= s2:
        continue

    inside = False
    for x, y in points + line_centres:
        if min_x < x < max_x and min_y < y < max_y:
            inside = True
            break

    if not inside:
        s2 = area

print(s1, s2)
```

# 10

```python
lamps, button_groups, joltages = [], [], []

for line in open("data/10.txt").read().splitlines():
    line = line.strip()
    lamp_str, rest = line.split("] ", 1)
    lamp = [0 if c == "." else 1 for c in lamp_str[1:]]
    btn_str, jolts_str = rest.split(" {", 1)

    buttons = []
    for b in btn_str.split():
        button = [0] * len(lamp)
        for i in b[1:-1].split(","):
            button[int(i)] = 1
        buttons.append(button)

    lamps.append(lamp)
    button_groups.append(buttons)
    joltages.append([int(x) for x in jolts_str[:-1].split(",")])
```

```python
def solve_integer_combo(a, B):
    n = len(a)
    m = len(B[0])
    prob = pulp.LpProblem("min_sum_x", pulp.LpMinimize)
    x = [pulp.LpVariable(f"x_{j}", lowBound=0, cat="Integer") for j in range(m)]
    prob += pulp.lpSum(x)
    for i in range(n):
        prob += pulp.lpSum(B[i][j] * x[j] for j in range(m)) == a[i]
    prob.solve(pulp.PULP_CBC_CMD(msg=False))
    return sum([int(v.value()) for v in x])
```

```python
s1 = s2 = 0
for lamp, buttons, joltage in zip(lamps, button_groups, joltages):
    for n in range(1, len(buttons) + 1):
        for comb in combinations(buttons, n):
            res = [sum(b[i] for b in comb) % 2 for i in range(len(lamp))]
            if res == lamp:
                s1 += n
                break
        else:
            continue
        break
    buttons = [list(row) for row in zip(*buttons)]
    s2 += solve_integer_combo(joltage, buttons)

print(s1, s2)
```

```python
inputs = open("data/11.txt").read().splitlines()
G = {inp.split(":")[0]: inp.split(":")[1].split() for inp in inputs}


def count_paths_dag(graph, start, target):
    reachable = set()
    q = deque([start])
    reachable.add(start)
    while q:
        u = q.popleft()
        for v in graph.get(u, []):
            if v not in reachable:
                reachable.add(v)
                q.append(v)

    if target not in reachable:
        return 0

    indeg = defaultdict(int)
    for u in reachable:
        indeg.setdefault(u, 0)
    for u in reachable:
        for v in graph.get(u, []):
            if v in reachable:
                indeg[v] += 1

    topo = []
    q = deque([u for u in reachable if indeg[u] == 0])
    while q:
        u = q.popleft()
        topo.append(u)
        for v in graph.get(u, []):
            if v not in reachable:
                continue
            indeg[v] -= 1
            if indeg[v] == 0:
                q.append(v)

    ways = {u: 0 for u in reachable}
    ways[target] = 1

    for u in reversed(topo):
        if u == target:
            continue
        ways[u] = sum(ways[v] for v in graph.get(u, []) if v in ways)

    return ways[start]


s1 = count_paths_dag(G, "you", "out")

a_b, b_c, c_d = (
    count_paths_dag(G, "svr", "fft"),
    count_paths_dag(G, "fft", "dac"),
    count_paths_dag(G, "dac", "out"),
)
a_c, c_b, b_d = (
    count_paths_dag(G, "svr", "dac"),
    count_paths_dag(G, "dac", "fft"),
    count_paths_dag(G, "fft", "out"),
)
s2 = a_b * b_c * c_d + a_c * c_b * b_d

print(s1, s2)
```
